```python
import xml.etree.ElementTree as ET

import requests

YIELD_XML_URL = (
    "https://home.treasury.gov/resource-center/data-chart-center/interest-rates/pages/xml"
    "?data=daily_treasury_yield_curve&field_tdr_date_value=all"
)


def fetch_yield_curve() -> dict[str, float]:
    """Fetches the latest daily treasury par yield curve from the XML feed.
    Returns a dict mapping term labels (e.g. "1 Mo", "2 Yr", etc) to yield floats.
    """
    resp = requests.get(YIELD_XML_URL)
    resp.raise_for_status()
    xml = resp.text

    root = ET.fromstring(xml)
    # The XML uses OData / Atom entry/field structure; you’ll need to navigate accordingly.
    # Example: find all <entry> tags, get the last one by date (or they may already be ordered).
    entries = root.findall(".//{http://www.w3.org/2005/Atom}entry")
    if not entries:
        raise RuntimeError("No entries in yield curve XML feed")

    # Assume entries are time-ordered; pick the last
    latest = entries[-1]
    fields = latest.findall(".//{http://schemas.microsoft.com/ado/2007/08/dataservices/metadata}properties")[0]

    curve = {}
    for child in fields:
        tag = child.tag
        text = child.text
        # tag might look like "d:BC_1MONTH" or "d:BC_10YEAR", etc. You'll need to strip namespaces
        if text is None:
            continue
        try:
            value = float(text)
        except ValueError:
            continue
        # Map tag to a nice label
        if "BC_1MONTH" in tag:
            curve["1M"] = value
        elif "BC_2MONTH" in tag:
            curve["2M"] = value
        elif "BC_3MONTH" in tag:
            curve["3M"] = value
        elif "BC_6MONTH" in tag:
            curve["6M"] = value
        elif "BC_1YEAR" in tag:
            curve["1Y"] = value
        elif "BC_2YEAR" in tag:
            curve["2Y"] = value
        elif "BC_3YEAR" in tag:
            curve["3Y"] = value
        elif "BC_5YEAR" in tag:
            curve["5Y"] = value
        elif "BC_7YEAR" in tag:
            curve["7Y"] = value
        elif "BC_10YEAR" in tag:
            curve["10Y"] = value
        elif "20YEAR" in tag:
            curve["20Y"] = value
        elif "30YEAR" in tag:
            curve["30Y"] = value
        # add other maturities if present
    return curve
```
